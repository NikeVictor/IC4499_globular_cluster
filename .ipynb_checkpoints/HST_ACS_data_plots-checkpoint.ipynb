```python
from astroquery.mast import Catalogs
from astropy.coordinates import SkyCoord
import astropy.units as u
import pandas as pd

coord = SkyCoord(ra=225.099, dec=-82.204, unit=(u.deg, u.deg), frame='icrs')

radius = 0.1 * u.deg
hsc_table = Catalogs.query_region(coordinates=coord, radius=radius, catalog="HSC")
df = hsc_table.to_pandas()

selected_columns = df[['MatchID', 'Distance', 'MatchRA', 'MatchDec', 'A_F606W', 'A_F606W_MAD', 
                            'A_F606W_N', 'A_F814W', 'A_F814W_MAD', 'A_F814W_N']]

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 10)

selected_columns
```

```python
import matplotlib.pyplot as plt

df['F606W_F814W_color'] = selected_columns['A_F606W'] - selected_columns['A_F814W']

plt.scatter(df['F606W_F814W_color'], selected_columns['A_F606W'], s=0.1)

plt.xlabel('F606W - F814W')  
plt.ylabel('F606W')
plt.title('color magnitude diagram')

plt.gca().invert_yaxis()
plt.xlim(-0.5, 3.0)
plt.savefig("cmd_plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

filename = 'grid-1.txt' 
data = np.loadtxt(filename, skiprows=1)

F606W = data[:, 10]
F814W = data[:, 15]

color = F606W - F814W

df['F606W_F814W_color'] = selected_columns['A_F606W'] - selected_columns['A_F814W']

plt.scatter(df['F606W_F814W_color'], selected_columns['A_F606W'], s=0.1, label='Original Data')
plt.scatter(color, F606W, s=0.1, color='red', label='Overlay Data')  # New overlay data in red

plt.xlabel('F606W - F814W')
plt.ylabel('F606W')
plt.title('Color Magnitude Diagram')

plt.gca().invert_yaxis()
plt.xlim(-0.5, 3.0)

plt.legend()
plt.savefig("overlay_plot_1")
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

df['F606W_F814W_color'] = selected_columns['A_F606W'] - selected_columns['A_F814W']

mean = np.mean(df['F606W_F814W_color'])
std_dev = np.std(df['F606W_F814W_color'])

x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)

y = norm.pdf(x, mean, std_dev) * len(df) * 0.3

plt.figure(figsize=(8, 6))

plt.hist(df['F606W_F814W_color'], bins=30, color='blue', alpha=0.7, edgecolor='black', density=False)

plt.plot(x, y, 'r-', lw=2, label='Gaussian fit')

plt.xlabel('F606W - F814W')
plt.ylabel('Number of Stars')
plt.title('IC4499')
plt.legend()

plt.savefig("color_index_histogram.png")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.coordinates import Distance

def read_dsed_isochrone(filename):
    data = pd.read_csv(filename, sep=r'\s+', comment='#')
    return data

age = 10  
metallicity = 0.02 

    
filename = 'grid-7.txt'  

isochrone_data = read_dsed_isochrone(filename)

distance_kpc = 40
distance = Distance(distance_kpc * u.kpc)
distmod = distance.distmod.value  
print(distmod)

try:
    mag_g = isochrone_data.iloc[:, 9] + distmod  
    mag_i = isochrone_data.iloc[:, 11] + distmod  

    color_g_i = mag_g - mag_i

    iso_df = pd.DataFrame({
        'mag_g': mag_g,
        'color_g_i': color_g_i
    })

    plt.figure(figsize=(10, 6))
    
    df['F606W_F814W_color'] = selected_columns['A_F606W'] - selected_columns['A_F814W']

    plt.scatter(df['F606W_F814W_color'], selected_columns['A_F606W'], s=0.1, label='Original Data')
    plt.scatter(iso_df['color_g_i'], iso_df['mag_g'], label='DSED Isochrone', color='red', s=1)
    plt.ylim(13, 25)
    plt.gca().invert_yaxis()  # Invert y-axis for CMD
    plt.xlim(-0.5, 3.0)
    plt.xlabel('Color (F606W - F814W)')
    plt.ylabel('Magnitude (F606W)')
    plt.title(f'DSED Isochrone Overlay on CMD (Age: {age} Gyr, [Fe/H]: {metallicity})')
    plt.show()

except KeyError as e:
    print(f"Column not found: {e}. Please check the available columns.")
```
